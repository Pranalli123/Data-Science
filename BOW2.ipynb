```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Step 1: Sample text (2 simple sentences)
texts = [
    "I like apples",
    "I like oranges"
]
```

```python
# Step 2: Create the CountVectorizer object
vectorizer = CountVectorizer()
```

```python
# Step 3: Learn the vocabulary and transform the texts into vectors
X = vectorizer.fit_transform(texts)
```

```python
# Step 4: Print the vocabulary (word to column index mapping)
print("Vocabulary:")
print(vectorizer.vocabulary_)
```

```python
# Step 5: Print the vectors (numerical representation of each sentence)
print("\nVectors:")
print(X.toarray())
```

```python
# Optional: Print feature names in order of columns
print("\nFeature names:")
print(vectorizer.get_feature_names_out())
```
