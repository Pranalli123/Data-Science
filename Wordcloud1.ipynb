```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

all_text = "good good good bad average nice nice"

# Generate word cloud
wordcloud = WordCloud().generate(all_text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
plt.imshow(wordcloud, interpolation="bilinear")      # displays the image smoothly.
plt.axis("off")  # hides the axes.
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Machine learning is fun and powerful learning learning machine"

wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color="white",
    colormap="viridis"  # Viridis gradient                     # Viridis is the name of a color map
).generate(text)

plt.imshow(wordcloud, interpolation="bilinear")               # display image smoothly
plt.axis("off")
plt.show()
```
