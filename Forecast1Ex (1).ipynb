```python
# Example: Forecasting Sales with Moving Average
import pandas as pd

# Past 7 days sales data
sales = [100, 120, 130, 150, 170, 160, 180]
days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]

df = pd.DataFrame({"Day":days, "Sales":sales})
print(df)

# Simple Moving Average Forecast (last 3 days avg)
forecast = sum(sales[-3:]) / 3                      # This takes the last 3 elements from the sales list.
                                                    # sales[-3:] = [170,160,180].
                                                    # Adds those last 3 values. sum=510/3=170
                                                    # so forcast is 170

print("Forecast for next day:", forecast)
```

```python
# Example 2: Forecasting Temperature using Last 2-Day Average
import pandas as pd

# Past 6 days temperature (°C)
temps = [30, 32, 31, 29, 28, 27]
days = ["Mon","Tue","Wed","Thu","Fri","Sat"]

df = pd.DataFrame({"Day":days, "Temperature":temps})
print(df)

# Simple Moving Average Forecast (last 2 days avg)
forecast = sum(temps[-2:]) / 2
print("Forecast for Sunday:", forecast)

```

```python
# Example 3: Forecasting Student’s Next Test Score
import pandas as pd

# Last 5 test scores
marks = [60, 65, 70, 68, 72]
tests = ["Test1","Test2","Test3","Test4","Test5"]

df = pd.DataFrame({"Test":tests, "Marks":marks})
print(df)

# Forecast next test score using last 2 tests average
forecast = sum(marks[-2:]) / 2
print("Forecast for Test6:", forecast)
```
