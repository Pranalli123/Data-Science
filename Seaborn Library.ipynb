```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    "Category": ["A", "B", "C", "D"],
    "Values": [10, 25, 18, 30]
}

# Create bar plot
sns.barplot(x="Category", y="Values", data=data)
plt.title("Simple Bar Chart using Seaborn")
plt.show()
```
