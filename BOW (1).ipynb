```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Step 1: Simple text data
texts = [
    "Cats like milk",
    "Dogs like meat"
]
```

```python
# Step 2: Create a CountVectorizer instance
vectorizer = CountVectorizer()
```

```python
# Step 3: Convert the text data to a bag of words (fit and transform)
X = vectorizer.fit_transform(texts)
```

```python
# Step 4: Show the vocabulary (word-to-index mapping)
print("Vocabulary (word -> column index):")
print(vectorizer.vocabulary_)
```

```python
# Step 5: Show the BoW matrix
print("\nBag of Words matrix (word counts):")
print(X.toarray())
```

```python
# Optional: Show feature (word) names
print("\nFeature names (columns):")
print(vectorizer.get_feature_names_out())
```
