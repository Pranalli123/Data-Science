```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Sample text
text = """
Data science is an interdisciplinary field that uses scientific methods,
processes, algorithms and systems to extract knowledge and insights from
structured and unstructured data. Data science is related to data mining,
machine learning and big data.
"""
```

```python
# Create the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
```

```python
# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.title("Word Cloud Example")
plt.show()
```

```python
# import the library
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# (basic text)
text = """
Python is a powerful programming language. Python is used in data science, web development,
machine learning, automation, and much more. Python is easy to learn and has a great community.
"""

# create the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud  'nearest' interpolation 
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='nearest')  # use 'nearest' because show the rough text
plt.axis('off')  # hide the axis
plt.title("Word Cloud using interpolation='nearest'")
plt.show()
```
