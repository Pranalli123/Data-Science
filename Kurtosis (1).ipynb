```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Example data: exam marks of students
marks = [50, 55, 60, 62, 65, 70, 72, 75, 80, 90]

```

```python
# Calculate kurtosis
print("Kurtosis:", kurtosis(marks))
```

```python
# Plot histogram with bins=5
plt.hist(marks, bins=5, color="skyblue", edgecolor="black")
plt.title("Histogram of Marks (bins=5)")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Plot histogram with bins=5
plt.hist(marks, bins=5, color="lightgreen", edgecolor="black")
plt.title("Histogram of Marks (bins=5)")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```
