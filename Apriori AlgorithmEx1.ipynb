```python
#  pip install mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
```

```python
# Step 2: Sample dataset
data = {'Bread': [1,1,0,1,1],
        'Milk': [1,1,1,1,0],
        'Butter': [0,1,1,1,0],
        'Eggs': [0,1,0,1,1]}

df = pd.DataFrame(data)
```

```python
# Step 3: Find frequent itemsets
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)    # shows actual item names in the output
print(frequent_itemsets)
```

```python
#Step 4: Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)      # confidence is â‰¥ 70%
print(rules[['antecedents','consequents','support','confidence','lift']])
```
